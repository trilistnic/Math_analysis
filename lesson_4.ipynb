### 2) Построить график функции:
$а) y = x^{2} + 4x + 3$

```python
x = np.linspace(-5, 5, 100)
y = x**2+4*x+3
# Построение графика
plt.title("y = x2 + 4x + 3") # заголовок
plt.xlabel("x") # ось абсцисс
plt.ylabel("y") # ось ординат
plt.grid()      # включение отображение сетки
plt.plot(x, y)  # построение графика
```

$б) y = −2 sin 3x$

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = -2*np.sin(3*x)
# Построение графика
plt.title("y = -2*sin(3x)") # заголовок
plt.xlabel("x") # ось абсцисс
plt.ylabel("y") # ось ординат
plt.grid()      # включение отображение сетки
plt.plot(x, y)  # построение графика
```

$в) y =|\{x\} − \frac{1}{2}|$

```python
x = np.linspace(-5, 5, 10000)
y = abs(x%1 - 0.5)
# Построение графика
plt.title("y = |{x} - 0.5|") # заголовок
plt.xlabel("x") # ось абсцисс
plt.ylabel("y") # ось ординат
plt.grid()      # включение отображение сетки
plt.plot(x, y)  # построение графика
```

$2)lim_{x→-1}\frac{x^{3}+x+2}{x^{3}+1}=[\frac{0}{0}]=lim_{x→-1}\frac{(x^3+1)+(x+1)}{x^3+1}=lim_{x→-1}\frac{1+\frac{x+1}{(x^3+1)}}{1}=lim_{x→-1}1+\frac{(x+1)}{(x+1)(x^2-x+1)}=lim_{x→-1}1+\frac{1}{x^2-x+1}=1+\frac{1}{3}=\frac{4}{3}$

$3)lim_{x→3}\frac{\sqrt{2x+3}-3}{\sqrt{x-2}-1}=[\frac{0}{0}]=lim_{x→3}\frac{(\sqrt{2x+3}-3)(\sqrt{2x+3}+3)(\sqrt{x-2}+1)}{(\sqrt{x-2}-1)(\sqrt{2x+3}+3)(\sqrt{x-2}+1}=lim_{x→3}\frac{(2x-6)(\sqrt{x-2}+1)}{(x-3)(\sqrt{2x+3}+3)}=lim_{x→3}2\frac{\sqrt{x-2}+1}{\sqrt{2x+3}+3}=\frac{2}{3}$

$4)lim_{x→0}\frac{1-cos(x)}{x^{2}}=[\frac{0}{0}]=lim_{x→0}\frac{2sin^2(\frac{x}{2})}{x^{2}}=lim_{\frac{x}{2}→0}\frac{2sin(\frac{x}{2})sin(\frac{x}{2})}{4\frac{x}{2}\frac{x}{2}}=\frac{2}{4}=\frac{1}{2}$

$5)lim_{x→0}x*ctg(x)=[0*∞]=lim_{x→0}\frac{x*cos(x)}{sin(x)}=lim_{x→0}\frac{cos(x)}{\frac{sin(x)}{x}}=lim_{x→0}\frac{cos(x)}{1}=\frac{1}{1}=1$
